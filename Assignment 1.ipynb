```python

import requests
```

```python
url='https://www.chittorgarh.com/report/ipo_report_listing_day_gain/98/?sort=IPO%5FListing%5FDatedesc&page=11'
response=requests.get(url)
```

```python
response.status_code
```

```python
page_contents=response.text
page_contents[:1000]
```

```python
from bs4 import BeautifulSoup
doc=BeautifulSoup(page_contents,'html.parser')
type(doc)
```

```python
table_parent1=doc.find_all('table')
#len(tr_parent1)
type(table_parent1)
```

```python
len(table_parent1)
table_parent1[1]
tr_child1=table_parent1[1].find_all('tr')
len(tr_child1)
```

```python
td_child1=tr_child1[0].find_all('th')
```

```python
td_child1
```

```python
def parse_document(tr_tag):
    td_tag=tr_tag.find_all('td')
    issuer_company=td_tag[0].find('a').text.strip()
    listing_date=td_tag[1].text.strip()
    issue_price=td_tag[2].text.strip()
    issue_price_Rs_Cr=td_tag[3].text.strip()
    QIB=td_tag[4].text.strip()
    NII=td_tag[5].text.strip()
    RII=td_tag[6].text.strip()
    EMP=td_tag[7].text.strip()
    TOTAL=td_tag[8].text.strip()
    Open_price=td_tag[9].text.strip()
    Low_price=td_tag[10].text.strip()
    High_price=td_tag[11].text.strip()
    close_price=td_tag[12].text.strip()
    percentage_change=td_tag[13].text.strip() 
    return{
        'Issuer Company':issuer_company,
        'Listing Date':listing_date,
        'Issue Price':issue_price,
        'Issue Price(Rs Cr)':issue_price_Rs_Cr,
        'QIB':QIB,
        'NII':NII,
        'RII':RII,
        'EMP':EMP,
        'TOTAL':TOTAL,
        'Open Price':Open_price,
        'Low Price':Low_price,
        'High Price':High_price,
        'Close Price':close_price,
        'Percentage Change':percentage_change
    }
```

```python
all_records = []

```

```python
for i in range(1,32):
    all_records.append(parse_document(tr_child1[i]))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(all_records)
df
```

```python
df.to_csv('Project-1')
```
